```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Admin\Downloads\archive (13)\DailyDelhiClimateTest.csv")
```

```python
sns.scatterplot(x=df['date'],y=df['meantemp'])
```

```python
x = df['meantemp'].ewm(alpha=0.9).mean()
x1 = df['meantemp'].ewm(alpha=0.7).mean()
x2 = df['meantemp'].ewm(alpha=0.5).mean()
x3 = df['meantemp'].ewm(alpha=0.1).mean()
```

```python
x
```

```python
df['ewma1'] = x
df['ewma2'] = x1
df['ewma3'] = x2
df['ewma4'] = x3
```

```python
plt.figure(figsize=(10,5))
plt.scatter(df["date"], df["meantemp"], color="blue", label="Mean Temp", alpha=0.6)
plt.plot(df["date"], df["ewma1"], color="red", linewidth=2, label="EWMA")
plt.plot(df["date"], df["ewma2"], color="orange", linewidth=2, label="EWMA")
plt.plot(df["date"], df["ewma3"], color="yellow", linewidth=2, label="EWMA")
plt.plot(df["date"], df["ewma4"], color="green", linewidth=2, label="EWMA")
```
