<h1>Preprocessing</h1>

```python
import pandas as pd
import numpy as np
import re
import string
```

<h2>1.Lower Casing

```python
df = pd.read_csv('/kaggle/input/d/lakshmi25npathi/imdb-dataset-of-50k-movie-reviews/IMDB Dataset.csv')
```

```python
df['review']=df['review'].str.lower()
```

<h2>2.Regular Expression

```python
def remove_html(text):
    p = re.compile('<.*?>')
    return p.sub(r'',text)
```

```python
df['review']=df['review'].apply(remove_html)
```

<h2>3.Remove Punctuation</h2>

```python
exclude = string.punctuation
```

```python
def remove_pun(text):
    for char in exclude:
        text = text.replace(char,'')
    return text
```

```python
def remove_pun_1(text):
    return text.translate(str.maketrans('','',exclude))
```

```python
df['review'] = df['review'].apply(remove_pun_1)
```

<h2>5.Stop Words</h2>

```python
from nltk.corpus import stopwords
```

```python
def remove_stopwords(text):
    new_text = []
    
    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return " ".join(x)
```

```python
df['review'].apply(remove_stopwords)
```

<h2>8.Stemming</h2>

```python
from nltk.stem.porter import PorterStemmer
```

```python
def stem_word(text):
    return " ".join([PorterStemmer().stem(word) for word in text.split()])
```
